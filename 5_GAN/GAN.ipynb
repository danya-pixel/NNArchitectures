```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("y_train.shape",y_train.shape)
print("x_train.shape",x_train.shape)
print("y_test.shape",y_test.shape)
print("x_test.shape",x_test.shape)
```

```python
from vizualize import show_batch
show_batch(x_train, y_train)
```

```python
from model import GAN
numbers = [2]

model = GAN(numbers, learning_rate=1e-3, decay_rate=1e-4, epochs=100)
loss_D, loss_G = model.train(x_train, y_train)
```

```python
plt.plot([i for i in range(len(loss_D))], loss_D)
plt.plot([i for i in range(len(loss_G))], loss_G)

plt.xlabel("# training steps")
plt.ylabel("training loss")
plt.legend(['Discriminator', 'Generator'])
plt.show()
```
